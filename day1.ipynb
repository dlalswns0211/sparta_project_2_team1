```python
print(df_copy.isnull().sum())
```

```python
# 불량품 관련 컬럼(Defects) 중 값이 '2'인 컬럼과 갯수 출력
defect_cols = df['Defects'].columns

result = {}
for col in defect_cols:
    count = (df[('Defects', col)] >= 2).sum()
    if count > 0:
        result[col] = count

if result:
    print("값이 2 이상인 Defects 컬럼 목록:")
    for col, count in result.items():
        print(f"  {col}: {count}개")
else:
    print("값이 2인 Defects 컬럼이 없습니다.")
```

```python
defect_cols = df_copy['Defects'].columns

# 'Defects' 대분류 아래의 모든 컬럼에 대해 2 이상의 값을 1로 변경
for col in defect_cols:
    # Defects 하위의 col 값 중 2 이상인 경우를 1로 치환
    df_copy.loc[df_copy[('Defects', col)] >= 2, ('Defects', col)] = 1
```

```python
# 불량품 관련 컬럼(Defects) 중 값이 '2'인 컬럼과 갯수 출력
defect_cols = df_copy['Defects'].columns

result = {}
for col in defect_cols:
    count = (df_copy[('Defects', col)] >= 2).sum()
    if count > 0:
        result[col] = count

if result:
    print("값이 2 이상인 Defects 컬럼 목록:")
    for col, count in result.items():
        print(f"  {col}: {count}개")
else:
    print("값이 2인 Defects 컬럼이 없습니다.")
```
