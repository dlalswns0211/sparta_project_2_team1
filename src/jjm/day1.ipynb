```python
### 이상치 탐색
# 센서 데이터(Sensor 변수)의 이상치 탐색 및 처리

#  17  (Sensor, Melting_Furnace_Temp)  7535 non-null   float64
#  18  (Sensor, Air_Pressure)          7535 non-null   float64
#  19  (Sensor, Air_Pressure_Min)      7535 non-null   int64  
#  20  (Sensor, Air_Pressure_Max)      7535 non-null   int64  
#  21  (Sensor, Coolant_Temp)          7535 non-null   float64
#  22  (Sensor, Coolant_Temp_Min)      7535 non-null   int64  
#  23  (Sensor, Coolant_Temp_Max)      7535 non-null   int64  
#  24  (Sensor, Coolant_Pressure)      7535 non-null   float64
#  25  (Sensor, Factory_Temp)          7445 non-null   float64
#  26  (Sensor, Factory_Temp_Min)      7445 non-null   float64
#  27  (Sensor, Factory_Temp_Max)      7445 non-null   float64
#  28  (Sensor, Factory_Humidity)      7445 non-null   float64
#  29  (Sensor, Factory_Humidity_Min)  7445 non-null   float64
#  30  (Sensor, Factory_Humidity_Max)  7445 non-null   float64

sensor_cols = [
    ('Sensor', 'Melting_Furnace_Temp'), ('Sensor', 'Air_Pressure'),
    ('Sensor', 'Air_Pressure_Min'), ('Sensor', 'Air_Pressure_Max'),
    ('Sensor', 'Coolant_Temp'), ('Sensor', 'Coolant_Temp_Min'),
    ('Sensor', 'Coolant_Temp_Max'), ('Sensor', 'Coolant_Pressure'),
    ('Sensor', 'Factory_Temp'), ('Sensor', 'Factory_Temp_Min'),
    ('Sensor', 'Factory_Temp_Max'), ('Sensor', 'Factory_Humidity'),
    ('Sensor', 'Factory_Humidity_Min'), ('Sensor', 'Factory_Humidity_Max')
]

'''
대칭분포이면서 이상치 없을 때 산술평균
비대칭분포이면서 이상치 있을 때 중앙값
범주형 데이터일 때 최빈값
이상치의 영향은 줄이면서도, 전체적인 데이터의 흐름을 반영하고 싶을 때 절사평균
각 항목이 전체에서 차지하는 중요도나 비중이 다를 때 가중평균
'''

def iqr(data):
    Q1 = data.quantile(0.25)
    Q2 = data.quantile(0.5)
    Q3 = data.quantile(0.75)
    iqr = Q3 - Q1
    iqr_lower = Q1 - (1.5*iqr)
    iqr_upper = Q3 + (1.5*iqr)
    outliers = data[(data<iqr_lower) | (data>iqr_upper)]
    return len(outliers)

def z_score_2(data):
    z_scores = (data - data.mean()) / data.std()
    z_outliers_2 = data[np.abs(z_scores)>2]
    return len(z_outliers_2)

def z_score_3(data):
    z_scores = (data - data.mean()) / data.std()
    z_outliers_3 = data[np.abs(z_scores)>3]
    return len(z_outliers_3)

results = []

for col in sensor_cols:
    target = df_copy[col]

    results.append({
        '변수': col[1],
        'IQR 이상치': iqr(target),
        'Z_Score_2': z_score_2(target),
        'Z_Score_3': z_score_3(target)
    })

result = pd.DataFrame(results)
print(result)
```

```python
# 결과를 담을 빈 리스트
group_results = []

# Product_Type별로 그룹화
for p_type, group_df in df_copy.groupby(('Process', 'Product_Type')):
    
    for col in sensor_cols:
        target = group_df[col]
        
        group_results.append({
            'Product_Type': p_type,
            '변수명': col[1],
            'IQR': iqr(target),
            'Z_Score_2': z_score_2(target),
            'Z_Score_3': z_score_3(target)
        })


df_outlier_by_type = pd.DataFrame(group_results)

print(f"\n[Product_Type 1 이상치 확인]")
type1_outliers = df_outlier_by_type[df_outlier_by_type['Product_Type'] == 1]
print(type1_outliers)

print(f"\n[Product_Type 2 이상치 확인]")
type2_outliers = df_outlier_by_type[df_outlier_by_type['Product_Type'] == 2]
print(type2_outliers)
```

```python
# 이상치 판단 함수
def is_iqr_outlier(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    iqr_val = Q3 - Q1
    lower = Q1 - 1.5 * iqr_val
    upper = Q3 + 1.5 * iqr_val
    return (data < lower) | (data > upper)

# 센서 컬럼들에 대해 이상치 여부(True/False) 계산
outlier_mask = df_copy[sensor_cols].apply(is_iqr_outlier)

# 'Has_Outlier' 파생컬럼 생성
df_copy['Has_Outlier'] = outlier_mask.any(axis=1)

print(df_copy['Has_Outlier'].value_counts())


# 분석용 데이터프레임 생성
analysis_df = pd.DataFrame({
    '이상치유무': df_copy['Has_Outlier'],
    '불량여부': df_copy['Is_Defective']
})

report = pd.crosstab(analysis_df['이상치유무'], analysis_df['불량여부'], normalize='index') * 100
report.columns = ['정상(%)', '불량(%)']

print("### 이상치 유무에 따른 불량 발생률 비교 ###")
display(report)
```
