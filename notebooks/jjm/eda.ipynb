```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic' 
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../../data/DieCasting_Quality_Preprocessed.csv',header=[0,1])
df.info()
```

```python
### 공정 데이터(Process 변수)와 불량 발생 간 관계 분석

#  0   (Process, id)                   7535 non-null   int64  
#  1   (Process, Product_Type)         7535 non-null   int64  
#  2   (Process, Shot)                 7535 non-null   int64  
#  3   (Process, Velocity_1)           7535 non-null   float64
#  4   (Process, Velocity_2)           7535 non-null   float64
#  5   (Process, Velocity_3)           7535 non-null   float64
#  6   (Process, High_Velocity)        7535 non-null   float64
#  7   (Process, Cylinder_Pressure)    7535 non-null   int64  
#  8   (Process, Rapid_Rise_Time)      7535 non-null   float64
#  9   (Process, Biscuit_Thickness)    7535 non-null   int64  
#  10  (Process, Clamping_Force)       7535 non-null   int64  
#  11  (Process, Cycle_Time)           7535 non-null   float64
#  12  (Process, Pressure_Rise_Time)   7535 non-null   float64
#  13  (Process, Casting_Pressure)     7535 non-null   int64  
#  14  (Process, Spray_Time)           7535 non-null   float64
#  15  (Process, Spray_1_Time)         7535 non-null   float64
#  16  (Process, Spray_2_Time)         7535 non-null   float64

# ====================================================================

## (Process, id), (Process, Product_Type), (Process, Shot)


## velocity(속도) -> (Process, Velocity_1), (Process, Velocity_2), (Process, Velocity_3), (Process, High_Velocity), (Process, Cycle_Time)
## pressure(압력) -> (Process, Cylinder_Pressure), (Process, Pressure_Rise_Time), (Process, Casting_Pressure), (Process, Clamping_Force), (Process, Rapid_Rise_Time)
## temperature(온도) -> (Process, Biscuit_Thickness), (Process, Spray_Time), (Process, Spray_1_Time), (Process, Spray_2_Time)

defect_base_names = [
    'Short_Shot', 'Bubble', 'Exfoliation', 'Blow_Hole', 'Stain', 
    'Dent', 'Deformation', 'Contamination', 'Impurity', 'Crack', 
    'Scratch', 'Buring_Mark', 'Inclusions'
]

# _1, _2 불량 통합 (둘 중 하나라도 있으면 해당 불량 발생으로 간주)
for base in defect_base_names:
    col1 = ('Defects', f'{base}_1')
    col2 = ('Defects', f'{base}_2')
    df[('Analysis', base)] = ((df[col1] > 0) | (df[col2] > 0)).astype(int)

velocity_columns = [('Process', 'Velocity_1'), ('Process', 'Velocity_2'), ('Process', 'Velocity_3'), ('Process', 'High_Velocity'), ('Process', 'Cycle_Time')]
pressure_columns = [('Process', 'Cylinder_Pressure'), ('Process', 'Pressure_Rise_Time'), ('Process', 'Casting_Pressure'), ('Process', 'Clamping_Force'), ('Process', 'Rapid_Rise_Time')]
temperature_columns = [('Process', 'Biscuit_Thickness'), ('Process', 'Spray_Time'), ('Process', 'Spray_1_Time'), ('Process', 'Spray_2_Time')]

analysis_defect_cols = [('Analysis', base) for base in defect_base_names]

# 상관관계 계산 및 시각화
def plot_relation(process_group, group_name):
    corr = df[process_group + analysis_defect_cols].corr()
    # 공정변수(행) x 불량항목(열)만 추출
    target_corr = corr.loc[process_group, analysis_defect_cols]
    
    plt.figure(figsize=(12, 6))
    sns.heatmap(target_corr, annot=True, cmap='RdBu_r', center=0)
    plt.title(f"Correlation: {group_name} vs Defects")
    plt.xticks(rotation=45)
    plt.show()

# 그룹별 실행
plot_relation(velocity_columns, "Velocity Group")
plot_relation(pressure_columns, "Pressure Group")
plot_relation(temperature_columns, "Temp/Spray Group")
```

```python
### 불량 유형별 주요 특징 시각화(박스플롯, 상관 행렬)

#  31  (Defects, Short_Shot_1)         7535 non-null   int64  
#  32  (Defects, Bubble_1)             7535 non-null   int64  
#  33  (Defects, Exfoliation_1)        7535 non-null   int64  
#  34  (Defects, Blow_Hole_1)          7535 non-null   int64  
#  35  (Defects, Stain_1)              7535 non-null   int64  
#  36  (Defects, Dent_1)               7535 non-null   int64  
#  37  (Defects, Deformation_1)        7535 non-null   int64  
#  38  (Defects, Contamination_1)      7535 non-null   int64  
#  39  (Defects, Impurity_1)           7535 non-null   int64  
#  40  (Defects, Crack_1)              7535 non-null   int64  
#  41  (Defects, Scratch_1)            7535 non-null   int64  
#  42  (Defects, Buring_Mark_1)        7535 non-null   int64  
#  43  (Defects, Inclusions_1)         7535 non-null   int64  
#  44  (Defects, Short_Shot_2)         7535 non-null   int64  
#  45  (Defects, Bubble_2)             7535 non-null   int64  
#  46  (Defects, Exfoliation_2)        7535 non-null   int64  
#  47  (Defects, Blow_Hole_2)          7535 non-null   int64  
#  48  (Defects, Stain_2)              7535 non-null   int64  
#  49  (Defects, Dent_2)               7535 non-null   int64  
#  50  (Defects, Deformation_2)        7535 non-null   int64  
#  51  (Defects, Contamination_2)      7535 non-null   int64  
#  52  (Defects, Impurity_2)           7535 non-null   int64  
#  53  (Defects, Crack_2)              7535 non-null   int64  
#  54  (Defects, Scratch_2)            7535 non-null   int64  
#  55  (Defects, Buring_Mark_2)        7535 non-null   int64  
#  56  (Defects, Inclusions_2)         7535 non-null   int64  

# 'Defects' 상위 그룹 아래에 있는 모든 하위 컬럼들을 리스트로 추출
defect_cols = df['Defects'].columns 

# _1(유형 1)과 _2(유형 2)를 분리
defects_1 = [col for col in defect_cols if col.endswith('_1')]
defects_2 = [col for col in defect_cols if col.endswith('_2')]

# 시각화: 이제 'Short_Shot_1', 'Bubble_1' 등이 개별적으로 표시됩니다.
plt.figure(figsize=(20, 6))
sns.boxplot(data=df['Defects'][defects_1]) # 반드시 ['Defects'] 그룹을 선택 후 하위 컬럼 투입
plt.xticks(rotation=45)
plt.title("Individual Defect Types Analysis (_1)")
plt.show()
plt.show()
```

```python
def plot_corr(columns, title):
    plt.figure(figsize=(12, 10))
    corr = df[columns].corr()
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='RdBu_r', center=0, linewidths=0.5)
    plt.title(f'Correlation Matrix: {title}', fontsize=15)
    plt.show()

# Type 1 상관관계
plot_corr(defects_1, "Defects Type 1 (_1)")

# Type 2 상관관계
plot_corr(defects_2, "Defects Type 2 (_2)")
```
