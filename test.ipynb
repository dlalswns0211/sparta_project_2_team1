1️⃣ 기본 정보

Shot: Shot 번호

id: 설비 ID

Product_Type: 제품 유형 (Type1, Type2)



2️⃣ 사출 및 압력 관련 변수

Velocity_1: 금형에 용탕을 투입하는 속도 (1단계)

Velocity_2: 금형에 용탕을 투입하는 속도 (2단계)

Velocity_3: 금형에 용탕을 투입하는 속도 (3단계)

High_Velocity: 고속 사출 속도

Cylinder_Pressure: Shot을 찍기 위한 실린더 압력

Casting_Pressure: Shot을 찍을 때의 압력

Clamping_Force: Shot을 찍을 때 금형을 밀어주는 힘

Rapid_Rise_Time: 실린더 압력 상승 시간

Pressure_Rise_Time: 주조 압력 상승 시간

Air_Pressure: 에어 압력 측정값

Air_Pressure_Min: 에어 압력 최소값

Air_Pressure_Max: 에어 압력 최대값



3️⃣ 시간 관련 변수

Cycle_Time: 제품이 나오기까지 걸리는 총 시간

Spray_Time: 스프레이 작업 시간

Spray_1_Time: 1차 스프레이 작업 시간

Spray_2_Time: 2차 스프레이 작업 시간



4️⃣ 온도 관련 변수

Melting_Furnace_Temp: 알루미늄을 녹여 일정 온도를 유지하는 장치 온도

Coolant_Temp: 냉각수 온도 측정값

Coolant_Temp_Min: 냉각수 온도 최소값

Coolant_Temp_Max: 냉각수 온도 최대값

Factory_Temp: 공장 온도 측정값

Factory_Temp_Min: 공장 온도 최소값

Factory_Temp_Max: 공장 온도 최대값



5️⃣ 습도 관련 변수

Factory_Humidity: 공장 습도 측정값

Factory_Humidity_Min: 공장 습도 최소값

Factory_Humidity_Max: 공장 습도 최대값



6️⃣ 기타 공정 변수

Biscuit_Thickness: 주조 비스켓 두께

Coolant_Pressure: 냉각수 압력 측정값



7️⃣ 불량 유형 (Cavity 1, 2 기준)

각 불량은 Cavity 1, 2에 대해 개별적으로 기록됨.

Short_Shot_1, 2: 미성형 불량

Bubble_1, 2: 기포 불량

Exfoliation_1, 2: 박리 불량

Blow_Hole_1, 2: 기공 불량

Stain_1, 2: 얼룩 불량

Dent_1, 2: 찍힘 불량

Deformation_1, 2: 평탄(변형) 불량

Contamination_1, 2: 오염 불량

Impurity_1, 2: 이물 불량

Crack_1, 2: 뜯김/균열 불량

Scratch_1, 2: 스크래치 불량

Buring_Mark_1, 2: 소착 불량

Inclusions_1, 2: 개재물 불량

```python
# 변수 'Biscuit_Thickness ', 'Clamping_Force ', ' Pressure_Rise_Time' 앞 뒤 공백 제거
df.columns = pd.MultiIndex.from_tuples(
    [(col[0].strip(), col[1].strip()) for col in df.columns]
)

print(df.columns.tolist())
```

```python
# factory_cols 결측치 중앙값 처리
factory_cols = [
    'Factory_Temp',
    'Factory_Temp_Min',
    'Factory_Temp_Max',
    'Factory_Humidity',
    'Factory_Humidity_Min',
    'Factory_Humidity_Max'
]

for col in factory_cols:
    df[('Sensor', col)] = df[('Sensor', col)].fillna(
        df[('Sensor', col)].median()
    )

df.isnull().sum().sum()
print("중앙값으로 처리 완료!")
```

```python
# High_Velocity, Rapid_Rise_Time, Pressure_Rise_Time가 모두 0인 행 4개 제거
zero_cols = ["High_Velocity", "Rapid_Rise_Time", " Pressure_Rise_Time"]

df = df[(df["High_Velocity"] != 0) |
        (df["Rapid_Rise_Time"] != 0) |
        (df[" Pressure_Rise_Time"] != 0)]
```

```python
# Factory_Temp, Factory_Humidity = Min, Max가 단순한 설정값이며 측정값은 Max를 아득히 상회함
print("측정값 개수 : ", len(df["Factory_Temp"]))
print("Factory_Temp < Min 개수:",
      (df["Factory_Temp"] < df["Factory_Temp_Min"]).sum())

print("Factory_Temp > Max 개수:",
      (df["Factory_Temp"] > df["Factory_Temp_Max"]).sum())

print("Factory_Humidity < Min 개수:",
      (df["Factory_Humidity"] < df["Factory_Humidity_Min"]).sum())

print("Factory_Humidity > Max 개수:",
      (df["Factory_Humidity"] > df["Factory_Humidity_Max"]).sum())

factory_cols = [
    "Factory_Temp_Min", "Factory_Temp_Max",
    "Factory_Humidity_Min", "Factory_Humidity_Max"
]
```

```python
# Air_Pressure, Coolant  = Min, Max가 단순한 설정값이며 측정값은 Min, Max 범위 내를 벗어나지 않음
print("측정값 개수 : ", len(df["Air_Pressure"]))
print("Air_Pressure < Min 개수:",
      (df["Air_Pressure"] < df["Air_Pressure_Min"]).sum())

print("Air_Pressure > Max 개수:",
      (df["Air_Pressure"] > df["Air_Pressure_Max"]).sum())


print("측정값 개수 : ", len(df["Coolant_Temp"]))
print("Coolant_Temp < Min 개수:",
      (df["Coolant_Temp"] < df["Coolant_Temp_Min"]).sum())

print("Coolant_Temp > Max 개수:",
      (df["Coolant_Temp"] > df["Coolant_Temp_Max"]).sum())

min_max_cols = [
    "Air_Pressure_Min", "Air_Pressure_Max",
    "Coolant_Temp_Min", "Coolant_Temp_Max"
]
```

```python
# 중앙값 처리
print("Factory_Temp 결측 행 수:",
      df["Factory_Temp"].isna().sum())

print("Factory_Humidity 결측 행 수:",
      df["Factory_Humidity"].isna().sum())

print("둘 중 하나라도 결측인 행 수:",
      df[["Factory_Temp", "Factory_Humidity"]].isna().any(axis=1).sum())

df = df.dropna(subset=["Factory_Temp", "Factory_Humidity"])
```

```python
drop_cols = factory_cols + min_max_cols

df_clean = df.drop(columns=drop_cols)

print("=== Drop 후 데이터 ===")
print("행 개수:", df_clean.shape[0])
print("컬럼 개수:", df_clean.shape[1])
```
